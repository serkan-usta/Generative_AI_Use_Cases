```python
import plotly.express as px
import pandas as pd
import numpy as np

# Define the data for the heatmap
industries = ['Banking and Securities', 'Communications', 'IT', 'Education', 'Government', 'Healthcare',
              'Insurance', 'Manufacturing', 'Pharma', 'Retail', 'Transportation', 'Utilities']
use_cases = ['Code', 'Video and image', 'Audio', 'Advertising solution', 'Text', 
             'AI avatar', 'Virtual assistant', 'Drug discovery', 'Search engine optimization', 'Data and analytics']

# Create a matrix of data (values can be adjusted to represent actual data)
values = np.array([
    [3, 2, 3, 1, 0, 2, 2, 3, 1, 2, 2, 1],
    [2, 3, 3, 1, 2, 2, 1, 2, 1, 3, 1, 2],
    [1, 1, 2, 1, 0, 2, 2, 3, 2, 1, 1, 1],
    [2, 2, 1, 3, 3, 1, 1, 2, 0, 3, 3, 1],
    [3, 3, 3, 1, 1, 2, 2, 3, 1, 3, 1, 2],
    [1, 0, 2, 3, 3, 1, 2, 3, 3, 1, 2, 1],
    [2, 2, 1, 1, 0, 3, 1, 2, 1, 2, 2, 1],
    [1, 3, 3, 2, 3, 3, 2, 3, 1, 3, 2, 3],
    [2, 1, 1, 3, 3, 1, 1, 2, 1, 3, 3, 1],
    [1, 2, 1, 1, 2, 2, 2, 2, 1, 3, 1, 1]
])

# Convert to DataFrame
df = pd.DataFrame(values, index=use_cases, columns=industries)

# Create the heatmap
fig = px.imshow(df, text_auto=True, color_continuous_scale='Blues', 
                title="Generative AI Use Cases by Industry", 
                labels=dict(x="Industries", y="Use Cases", color="Frequency"))

# Update layout for better display
fig.update_layout(
    xaxis=dict(title="Industry", tickangle=45),
    yaxis=dict(title="Use Case"),
    autosize=True
)

# Save the heatmap as an HTML file
fig.write_html('interactive_heatmap.html')

# Display the figure in Kaggle
fig.show()
```
